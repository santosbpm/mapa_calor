```python
import folium
import pandas as pd
from folium.plugins import HeatMap


url = (
    "https://raw.githubusercontent.com/santosbpm/mapa_calor/master"
)

# Dados para o HeatMap
dados_calor = pd.read_csv(f"{url}/dados_calor.csv")

# Dados com a área do município
geoMunicipio = f"{url}/limites_municipio.geojson"

# Dados com a área do estado e seus municípios
geoRj = f"{url}/limites_estado.json"

# Cria um mapa
mapa_rj = folium.Map(location=[-22.8972131, -43.5702964], zoom_start=9)

# Demarca o mapa utilizando os dados anteriores
folium.GeoJson(geoRj).add_to(mapa_rj)
folium.GeoJson(geoMunicipio, style_function=lambda feature: {
        'fillColor': 'red',
        'color': 'red'
    }
).add_to(mapa_rj)

# Adicionar o HeatMap ao mapa
HeatMap(dados_calor).add_to(mapa_rj)

folium.Marker([-22.90196, -43.57932], popup="UERJ-ZO - Universidade do Estado do Rio de Janeiro campus Zona Oeste").add_to(mapa_rj)

mapa_rj
```
